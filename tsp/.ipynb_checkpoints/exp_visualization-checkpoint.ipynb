```python
import matplotlib.pyplot as plt
import os
import json
import numpy as np
```

```python
results = []
for exp in range(26):
    root_dir = "./logs/exp%d/1"%exp
    file = os.path.join(root_dir,'logs.txt')

    f = open(file)
    init_time = float(f.readline().split(",")[1][:-1])
    result = []
    for i in range(500):
        a = f.readline()
        itr,t,v = a.split(', ')
        result.append((int(itr),float(t),float(v[:-1])))

    args = json.load(open(os.path.join(root_dir,'args.txt')))
    results.append((result,args,init_time))
```

```python
greedy_ratio_axis = []
num_clusters_axis = []
values = []
times = []
for result in results:
    arg = result[1]
    greedy_ratio_axis.append(arg['greedy_ratio'])
    num_clusters_axis.append(arg['num_clusters'])
    values.append(result[0][-1][-1])
    time = result[2]
    for i in range(500):
        time += result[0][i][1]
    times.append(time)
```

```python
plt.scatter([-np.log(1-v) for v in greedy_ratio_axis[:-1]],[np.log(v) for v in num_clusters_axis[:-1]],c = [np.log(v) for v in values[:-1]],s = 200)
plt.title('relationship between hyperparameter and tsp distance')
plt.xlabel('log-scale greedy ratio')
plt.ylabel('log-scale num clusters')
plt.colorbar(label = 'log scale tsp_distance')

plt.savefig('./images/hps-tsp_distance.png')
```

```python
plt.scatter([-np.log(1-v) for v in greedy_ratio_axis[:-1]],[np.log(v) for v in num_clusters_axis[:-1]],c = [np.log(v) for v in times[:-1]],s = 200)
plt.title('relationship between hyperparameter and execution time')
plt.xlabel('log-scale greedy ratio')
plt.ylabel('log-scale num clusters')
plt.colorbar(label = 'log scale time')
plt.savefig('./images/hps-time.png')
```
